Salah satu sektor yang sangat bergantung pada bahan bakar minyak adalah transportasi. Selanjutnya, emisi CO2 dari sektor transportasi menyumbang 22,9% dari emisi CO2 global. Antara tahun 2000 dan 2010, jumlah kendaraan meningkat 15,1% per tahun, meningkat dari 18,98 juta pada tahun 2000 menjadi 77,13 juta pada tahun 2010. Pengurangan bahan bakar fosil di masa depan memotivasi para peneliti, perancang, produsen mobil, dan agen transportasi untuk mencari sumber daya alternatif, seperti listrik, untuk kendaraan.
